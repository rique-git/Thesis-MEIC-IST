## Encoding Categorical Variables

```python
def encoding(df):
    df["Heart Disease"] = df["Heart Disease"].map({"Absence":0, "Presence": 1})
    df = df.rename(columns={'EKG results': 'ECG results'})
    df = df[df['ECG results'] != 1]

    return df
```

```python
data = encoding(data)
```

## Normalization
scale and standardize

```python
def scale_and_standardize(df, target):
    X = df.drop(columns=[target])
    y = df[target]

    # Scale to [0, 1]
    scaler_minmax = MinMaxScaler()
    X_minmax = scaler_minmax.fit_transform(X)

    # Standardize to zero mean and unit variance
    scaler_std = StandardScaler()
    X_std = scaler_std.fit_transform(X_minmax)

    # Convert back to DataFrame
    X_scaled_std = pd.DataFrame(X_std, columns=X.columns, index=df.index)

    # Combine with target
    df_final = pd.concat([X_scaled_std, y], axis=1)

    return df_final

data = scale_and_standardize(data, "Heart Disease")
```
